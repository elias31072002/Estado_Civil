```python
file = r'C:\Users\iasfa\Desktop\Dados_Públicos\perfil_eleitorado_ATUAL.csv'
df = pd.read_csv(file, delimiter=';',
                 error_bad_lines=False, encoding='latin1')
```

```python

df.groupby('QT_ELEITORES_PERFIL')['DS_ESTADO_CIVIL'].max().sort_values(ascending=False)
```

```python
df['ESTADO_CIVIL_EST'] = df['SG_UF']+' '+df['DS_ESTADO_CIVIL']
df.groupby('ESTADO_CIVIL_EST').size().sort_values().tail(10).plot(kind='barh', color='GREEN',  figsize=(15, 9), title='Estado civil por estado nas eleições municipais de 2020')
```

```python
df['ESTADO_CIVIL_FE'] = df['DS_FAIXA_ETARIA']+' '+df['DS_ESTADO_CIVIL']
jcr = df.query('NM_MUNICIPIO == "JACAREÍ"')
jcr.tail(30).plot(kind='bar', figsize=(20,13), title="Estado civil dos eleitores na cidade de Jacareí",
        x= 'ESTADO_CIVIL_FE', y='QT_ELEITORES_PERFIL', color='PINK')
```
